EL JUEGO DEL AHORCADO POR NAVIDAD
**¿Cuáles son sus premisas?**
 - La jugadora debe adivinar la palabra oculta que estará representada por guiones bajos.
 - La jugadora irá introduciendo letras. Si la letra introducida se encuentra en la palabra oculta, esta letra se descubrirá. Por el contrario, si la letra introducida no se encuentra en la palabra oculta, ha cometido un fallo y se dibujará una parte del ahorcado.
 - El juego finaliza cuando la jugadora descubre la palabra oculta sin ser "ahorcada" o si es "ahorcada" antes de descubrir la palabra.

 **Reglas del juego**
 - Se descarta el uso de tildes y la Ñ.
 - Las letras introducidas se mostrarán para evitar repeticiones y penalizaciones.
 - El número total de intentos es 7
 - Se hará uso de letras mayúsculas.

 **¿Qué necesitamos definir o tener en cuenta**
 - Un mensaje de bienvenida a las jugadoras.
 - La palabra oculta procedente de una lista de palabras navideñas.
 - La lista de palabras navideñas objeto del juego.
 - La petición a la jugadora de que introduzca una letra.
 - El tablero de juego [_ _ _ _ _] que se imprimirá por pantalla y que ha de ser igual a la longitud de la palabra oculta.
 - El progreso que este tablero irá experimentando a lo largo del juego, así como el dibujo del ahorcado.
 - Un registro de las letras introducidas/intentadas.
 - Un contador, que será el número de intentos, y que disminuirá cada vez que se falla.
 - El BUCLE PRINCIPAL.

EL BUCLE PRINCIPAL DEL JUEGO
¿Qué condiciones tiene este bucle? ¿Mientras existan qué escenarios se puede jugar?:
Mientras nos queden intentos y aún haya letras por descubrir.

1. Mostrar el estado actual del juego.
2. Pedir la entrada de la jugadora (mientras que los intentos sean mayor que cero, habrá que mostrar el estado y seguir pidiendo a la jugadora que introduzca una letra).
3. Comprobar la validez de esta entrada.
4. Comprobar si ya había intentado con ella.
5. Agregar la letra al registro de letras introducidas/intentadas.
6. Determinar si ha sido acierto o fallo.
7. Mostrar el resultado final 

```python
# Diseño y muestra del mensaje de bienvenida
print("Bienvenida al juego del Ahorcado por Navidad.")
print("Tu objetivo es descubrir la palabra navideña oculta antes de ser ahorcada.")
print("Para ello, has de introducir una letra cada vez que se te pida y afinar tu puntería ;).")
```

```python
# Definción de la lista de palabras navideñas
palabras_navideñas = ["ARBOL", "MAZAPAN", "ELFO", "TURRON", "TRINEO", "REGALO", "ESTRELLA", "POLVORON",
                      "VILLANCICO", "ZAMBOMBA", "MUERDAGO", "DUENDE", "NIEVE", "BELEN"]
```

```python
''' Definición de la palabra oculta y del tablero. Para la elección de la palabra oculta empleamos
la función random.choice que escogerá una palabra al azar de la lista creada en la celda anterior '''
import random
palabra_oculta = random.choice(palabras_navideñas)
tablero = ["_"]*len(palabra_oculta) #Creamos un tablero de la misma longitud que la palabra oculta
print(" ".join(tablero))
```

```python
# Definición del número de intentos y su registro
intentos_totales = 7 #Establecemos el máximo de intentos
intentos_restantes = intentos_totales #Inicializamos nuestro contador
```

```python
#Definición de la estructura en la que se almacenarán las letras introducidas
letras_jugadas = set()
#Escogemos un set porque no permite valores repetidos
```

```python
# Comenzamos con el diseño del bucle principal.
# 1. Muestra del estado actual del juego
print(f"Tu palabra oculta es : {" ".join(tablero)}")
print(f"Te quedan :{intentos_restantes} intentos")
print(f"Las letras ya utilizadas son: {",".join(sorted(list(letras_jugadas)))} ")
'''En esta última línea de código, lo que estamos haciendo es convertir el set de letras
en una lista para poder ordenarla y mostrar sus elementos por pantalla'''
```
